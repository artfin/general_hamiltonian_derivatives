```python
tmax = 2500000.0
sampling_time = 200.0
#ps_point = PS_point(23.0, -2.00393, 0.753412, -5.1339, 3.64236, 1.3157, 0.487732, 2.326, 0.979626, 2.99349) 
ps_point = PS_point.from_str('7.96108 3.46359 1.83784 -2.48123 1.03983 8.25804 2.74308 -27.2 6.14174 -4.35758')


time, phase_points = compute_trajectory(tmax, sampling_time, ps_point)

time_r, phase_points_r = compute_trajectory(tmax, sampling_time, reverse_phase_point(phase_points[-1]))
phase_points_r.reverse()
```

```python
%%time
energy = make_energy_array(time, phase_points, step=10)
energy_r = make_energy_array(time_r, phase_points_r, step=10)
```

```python
LineStyle = namedtuple('LineStyle', 
    ['data', 'color', 'linestyle', 'lw', 'alpha', 'name'],
    defaults = [None, 'k', '-', 1.0, 1.0, None] 
)
```

```python
plotter = TrajectoryPlotter(time, phase_points, figsize=(16,12))

R = stack_arrays(time, phase_points, 'R')
R_r = stack_arrays(time_r, phase_points_r, 'R')

plotter.set_locators({'xaxis.major' : 5.0, 'xaxis.minor': 2.5, 'yaxis.major': 1.0, 'yaxis.minor': 0.5})

plotter.set_data_to_plot([
    LineStyle(data = R, linestyle = '-'),
    LineStyle(data = R_r, linestyle = '-', color = 'r'),
])

xaxis_factor = 1e5
plotter.set_axis_limits(xlim=(0, tmax/xaxis_factor), ylim = (5.8, 10.2))
plotter.make_plot(xaxis_factor = xaxis_factor)

plt.show()
```
